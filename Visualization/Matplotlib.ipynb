# Why to visualize data?
- Data visualization is the practice of translating information into a visual context, such as a map or graph, to make data easier for the human brain to understand and pull insights from. The main goal of data visualization is to make it easier to identify patterns, trends and outliers in large data sets.
    - Better analysis
    - Quick action
    - Identifying patterns
    - Finding errors
    - Understanding the story
    - Exploring business insights
    - Grasping the Latest Trends

- Xlim: used to define limit for x variable 

- Ylim:used to define limit for y variable 

```python
area = [1000,1520,2000,3000,5210]
price = [10,15,12,25,30]
plt.plot(area,price)
plt.grid()
plt.xlim(1000,6000)
plt.ylim(5,50)
plt.xlabel('Area of the house')
plt.ylabel('Price of the house')
```

- Xticks : rotate label 

- Yticks: rotate label 

```python
area = [1000,1520,2000,3000,5210]
price = [10,15,12,25,30]
plt.plot(area,price)
plt.xticks(rotation = 45)
plt.yticks(rotation = 10)
plt.xlabel('Area of the house')
plt.ylabel('Price of the house')
```

- Multiple Plots: multiple line in one chart called multiple plot 

 - Legend : gives label to plots 

 - Color : used to give colors to plot

```python
x = np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)

plt.plot(x,y1,color = 'b', label = 'Square root')
plt.plot(x,y2, color = 'g',label = 'Log')
plt.legend()
```

# Matplolib Histogram 
- A histogram is a graph showing frequency distributions.

- It is a graph showing the number of observations within each given interval.

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Automobile_data.csv')
df.head()
```

```python
df.shape[0]
```

```python
df.describe().T
```

```python
plt.hist(df['width'], bins=10)
plt.title('Width frequency table')
plt.xlabel('Width')
plt.ylabel('Frequency')
```

```python
width_mean = df['width'].mean()
width_median = df['width'].median()

df['width'].hist(bins=10)
plt.axvline(width_mean, color = 'y')
plt.axvline(width_median, color = 'r')
plt.ylabel('Frequency')
plt.xlabel('Width')
```
